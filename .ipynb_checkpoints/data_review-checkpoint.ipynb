```python
import pandas as pd
import numpy as np
```

```python
# 读取 CSV 文件
df = pd.read_csv(r".\问题\Wimbledon_featured_matches.csv")
player_name = set(df["player1"].unique().tolist() + df["player2"].unique().tolist())
```

```python
most_common_elements = {}  # 创建一个空字典以存储每列的最常见元素

for name in player_name:
    # 根据条件筛选出满足条件的行
    condition = ((df["player1"] == name)) | ((df["player2"] == name))
    
    # 填充指定条件下的数据
    for col in df.columns:
        if col in most_common_elements:  # 如果列名已在字典中，则跳过
            continue

        value_counts = df.loc[condition, col].value_counts()
        if not value_counts.empty:
            most_common_elements[col] = value_counts.idxmax()
        else:
            most_common_elements[col] = -1  # 或者其他您认为合适的默认值

    # 直接修改满足条件的数据
    df.loc[condition] = df.loc[condition].fillna(most_common_elements)
```

```python
reflect_width = {'BC':1, 'B':2, 'BW':4, 'W':4, 'C':3}
reflect_serve_depth = {"CTL":2, "NCTL":1}
reflect_return_depth = {"D":2,"ND":1}

# pd.set_option('future.no_silent_downcasting', True)
df['serve_width'] = df['serve_width'].replace(reflect_width)
df['serve_depth'] = df['serve_depth'].replace(reflect_serve_depth)
df['return_depth'] = df['return_depth'].replace(reflect_return_depth)
df['if_ace'] = (df["p1_ace"] | df["p2_ace"]).astype(int)

# df[df["point_victor"]==2] = 0
# print(df["point_victor"])
# df['group'] = (df['server'] != df['server'].shift()).cumsum()
# # 对每个组应用聚合函数
# aggregated = df.groupby('group').agg(lambda x: x.mean() if np.issubdtype(x.dtype, np.number) else x.iloc[0]).reset_index(drop=True)
# # 删除辅助列并转换结果为列表形式
# result = [aggregated[col].tolist() for col in aggregated.columns if col != 'group']
# print(result)
# dic = { df.columns[i]:result[i] for i in range(len(result))}
# print(dic)
# df = pd.DataFrame(dic)
# print(df)

coids = list(df["match_id"].unique())
all_competition = []

for coid in coids:
    dft = df[df["match_id"] == coid]
    dp = pd.DataFrame(dft)

    # 选取需要进行计算的列
    selected_columns = ["p1_distance_run", "rally_count"]
    selected_columns_2 = ["p2_distance_run", "rally_count"]

    # 使用apply函数将这些列的数据进行计算并存入新的列"move_1"
    dp["move_1"] = dft[selected_columns[0]] #.apply(lambda row: row["p1_distance_run"] / row["rally_count"] if row["rally_count"] != 0 else 0, axis=1)
    dp["move_2"] = dft[selected_columns_2[0]] #.apply(lambda row: row["p2_distance_run"] / row["rally_count"] if row["rally_count"] != 0 else 0, axis=1)
    dp["p1_points_won"] = dft["p1_points_won"]
    dp["p2_points_won"] = dft["p2_points_won"]
    
    all_competition.append(dp)
```

```python
import matplotlib.pyplot as plt

for idx,dp in enumerate(all_competition):
    # dp['win1_mean'] = dp["point_victor"]
    # dp['win2_mean'] = 1-dp["point_victor"]
    ws = 6
    windows1 = dp['p1_points_won'].rolling(window=ws)
    times_mean = (windows1.max() - windows1.min())/(ws-1)
    all_competition[idx]["win1_mean"] = times_mean
    windows2 = dp['p2_points_won'].rolling(window=ws)
    times_mean = (windows2.max() - windows2.min())/(ws-1)
    all_competition[idx]["win2_mean"] = times_mean
    
```

```python
player1_shot = []

# 定义一个函数，该函数接受一行，并返回第一个非零值
def find_first_non_zero(row):
    # 列名顺序从'win1_6_mean'到'win1_3_mean'
    cols = ["win1_5_mean", "win1_4_mean", "win1_3_mean", "win1_2_mean"]
    if "win1_5_mean" not in row:
        cols = ["win2_5_mean", "win2_4_mean", "win2_3_mean", "win2_2_mean"]
    
    for col in cols:
        if row[col] != 0:
            return row[col]
    return 0  # 如果所有选定列都是0，返回NaN

insprice = np.array([0.041, 0.039, -0.03, -0.02])
for idx, dp in enumerate(all_competition):
    temp = dp.copy()
    temp.loc[:, ["win1_2_mean", "win1_3_mean", "win1_4_mean", "win1_5_mean"]] = (all_freWIn[idx][["win1_2_mean", "win1_3_mean", "win1_4_mean", "win1_5_mean"]].values.astype(int)*insprice)
    temp.loc[:, "win_inspire"] = temp[["win1_2_mean", "win1_3_mean", "win1_4_mean", "win1_5_mean"]].apply(find_first_non_zero, axis=1)
    temp["win_inspire"] = temp["win_inspire"].shift(1)
    player1_shot.append(temp)

temp.shape[0]
```

```python
player1_shot_win_ave = []
player_hit_win_ave = []
for pd in player1_shot:
    player1_shot_win_ave.append(pd[(pd["point_victor"] == 1) & (pd["server"] == 1)].shape[0]/pd.shape[0])
    player_hit_win_ave.append(pd[(pd["point_victor"] == 1) & (pd["server"] == 2)].shape[0]/pd.shape[0])
```

```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
import math

def normalize_series(s):
    return (s - s.min()) / (s.max() - s.min())

# 假设你的模型是 p + a*b + t*b + e*c + k*n，其中a, t, c, k是你需要拟合的参数
def model_func(xdata, a, c, k, m):
    p = xdata[:, 0]  # 假设第一列是p
    b = xdata[:, 1]  # 第二列是b
    e = xdata[:, 2]  # 第三列是e
    n = xdata[:, 3]  # 第四列是n
    # arr = np.arange(1, n.shape[0]+1)
    return p + a*b + c*e + k*n + m
```

```python
# 假设temp, player1_shot_win_ave已经正确定义
mean_acc = 0
mean_r = 0
pred_res = []
shot_acc_param = []
for i in range(len(player1_shot)):
    temp = player1_shot[i][(player1_shot[i]["server"] == 1)].copy().dropna()
    temp["p_m"] = player1_shot_win_ave[i] 
    col_name = ["ss","emotion_score", "win_inspire"]
    for n in col_name:
        temp[n] = normalize_series(temp[n])


    xdata = np.column_stack((temp["p_m"], temp["ss"], temp["emotion_score"], temp["win_inspire"]))
    ydata = temp["win1_mean"].values
    # 进行曲线拟合
    params, params_covariance = curve_fit(model_func, xdata, ydata)
    
    # 使用拟合后的参数和模型函数计算预测值
    sy_pred = model_func(xdata, *params)
    
    
    pred_res.append(np.column_stack([np.array(temp.index), np.array([sy_pred]).squeeze()]))
    
    ss_res = np.sum((ydata - sy_pred) ** 2)  # 残差平方和
    ss_tot = np.sum((ydata - np.mean(ydata)) ** 2)  # 总平方和
    r_squared = 1 - (ss_res / ss_tot)

    sydata = ydata > 0.5
    sy_pred = sy_pred > 0.5

    # 计算R²
    accuracy = (sydata == sy_pred).mean()
    
    shot_acc_param.append({"param":params,"acc":accuracy, "r2":r_squared})
    mean_acc += accuracy
    mean_r += r_squared

mean_acc /= len(player1_shot)
mean_r /= len(player1_shot)
# mean_acc, shot_acc_param
# pred_res
```

```python
# 假设temp, player1_shot_win_ave已经正确定义
mean_acc = 0
mean_r = 0
hit_acc_param = []
for i in range(len(player_hit_win_ave)):
    temp = player1_shot[i][(player1_shot[i]["server"] == 2)].copy().dropna()
    temp["p_m"] = player_hit_win_ave[i]  # 确保这行代码正确执行，player1_shot_win_ave[0] 应当是个具体的数值或者序列
    col_name = ["ss","emotion_score","win_inspire"]
    for n in col_name:
        temp[n] = normalize_series(temp[n])

    xdata = np.column_stack((temp["p_m"], temp["ss"], temp["emotion_score"], temp["win_inspire"]))
    ydata = temp["win1_mean"].values
    # 进行曲线拟合
    params, params_covariance = curve_fit(model_func, xdata, ydata)
    
    # 使用拟合后的参数和模型函数计算预测值
    sy_pred = model_func(xdata, *params)
    pred_res[i] = np.vstack([pred_res[i], np.column_stack([np.array(temp.index), np.array([sy_pred]).squeeze()])])
    sorted_indices = np.argsort(pred_res[i][:, 0])
    pred_res[i] = pred_res[i][sorted_indices]
    
    ss_res = np.sum((ydata - sy_pred) ** 2)  # 残差平方和
    ss_tot = np.sum((ydata - np.mean(ydata)) ** 2)  # 总平方和
    r_squared = 1 - (ss_res / ss_tot)
    sse = np.sum((ydata - sy_pred) ** 2)
    rmse = np.sqrt(np.mean((ydata - sy_pred) ** 2))

    sydata = ydata > 0.5
    sy_pred = sy_pred > 0.5

    # 计算R²
    accuracy = (sydata == sy_pred).mean()
    
    hit_acc_param.append({"param":params,"acc":accuracy, "r2":r_squared})
    mean_acc += accuracy
    mean_r += r_squared
    
mean_acc /= len(player1_shot)
mean_r /= len(player1_shot)
```

```python
comp_num = 1

temp = player1_shot[comp_num].copy().dropna()
ydata = temp["win1_mean"].values
w2data = temp["win2_mean"].values
player1_point = temp["p1_points_won"].values
player2_point = temp["p2_points_won"].values
y_pred = pred_res[comp_num][:,1]
# v1_sum = np.cumsum(ydata-0.5)/3
# v2_sum = np.cumsum(w2data-0.5)/3

def process_vo(data):
    v1_sum = np.zeros(data.shape)
    for i in range(len(data)):
        # 对于第i个数据点，计算从0到i的权重
        weights_for_i = 1.0 ** (-np.arange(len(data)-i-1, -1, -1))
        # 计算加权和
        weighted_sum_for_i = np.sum(data[:len(data)-i] * weights_for_i)
        # # 存储结果
        v1_sum[len(data)-i-1] = weighted_sum_for_i
    return v1_sum
v1_sum = process_vo(0.5-ydata)


ts = np.diff(ydata)

ydata -= 0.5
y_pred -= 0.5
residual_std = np.std(ydata - y_pred)
delta = 0.8 * residual_std  # 举例，使用95%置信区间
# 计算每个预测值的上下界
lower_bounds = y_pred - delta
upper_bounds = y_pred + delta
# 判断实际y值是否落在预测的范围内
within_range = (ydata >= lower_bounds) & (ydata <= upper_bounds)
# 计算落在范围内的比率
within_range_ratio = np.mean(within_range)

print("Ratio of actual y values within the predicted range:", within_range_ratio)
```

```python
from NetWorkPackage.MLP import MLP
from NetWorkPackage.Transformer import TransformerDecoder, TransformerDecoderLayer, generate_square_subsequent_mask
import pandas as pd
import torch
import torch.nn.functional as F
import torch.nn as nn 
import torch.optim as optim

def pad_and_create_mask(input_tensor, max_length):
    """
    将输入Tensor填充到max_length*14的大小，并生成对应的mask。
    
    参数:
    - input_tensor: 输入的Tensor，大小为n*14。
    - max_length: 填充后的目标长度。
    
    返回:
    - padded_tensor: 填充到max_length*14的Tensor。
    - mask: 对应的mask，标记实际数据位置为1，填充位置为0。
    """
    n,feature_dim = input_tensor.shape
    # 检查是否需要填充
    if n >= max_length:
        # 如果输入Tensor的长度已经大于或等于max_length，则不进行填充
        padded_tensor = input_tensor[:max_length]
        mask = torch.ones(max_length, dtype=torch.float32)
    else:
        # 计算需要填充的长度
        padding_length = max_length - n
        # 创建一个新的Tensor进行填充
        padded_tensor = F.pad(input_tensor, (0, 0, 0, padding_length), "constant", 0)
        # 生成mask
        mask = torch.cat([torch.zeros(n), torch.ones(padding_length)], dim=0)
    return padded_tensor, mask.to(torch.bool), n


insprice = np.array([1, 2, 3, 4, 5])
train_data = [pd.DataFrame() for i in range(len(all_competition))]
for idx, dp in enumerate(train_data):
    all_competition[idx].loc[all_competition[idx]["server"] == 2, "server"] = -1
    
    dp["ace"] = competition_emotion[idx]["ace"]
    dp["winner"] = competition_emotion[idx]["winner"]
    dp["double_fault"] = competition_emotion[idx]["double_fault"]
    dp["unf_err"] = competition_emotion[idx]["unf_err"]
    dp["break_pt_won"] = competition_emotion[idx]["break_pt_won"]
    dp["break_pt_missed"] = competition_emotion[idx]["break_pt_missed"]
    dp["net_pt"] = competition_emotion[idx]["net_pt"]
    dp["net_pt_won"] = competition_emotion[idx]["net_pt_won"]
    dp["server"] = all_competition[idx]["server"]
    dp["run"] = all_competition[idx]["p1_distance_run"] - all_competition[idx]["p2_distance_run"]
    dp["run"] = dp["run"] / dp["run"].abs().max()
    all_competition[idx].loc[:, ["win1_2_mean","win1_3_mean", "win1_4_mean", "win1_5_mean", "win1_6_mean"]] = (all_freWIn[idx][["win1_2_mean","win1_3_mean", "win1_4_mean", "win1_5_mean", "win1_6_mean"]].values.astype(int)*insprice)
    dp.loc[:, "all_win"] = all_competition[idx][["win1_2_mean","win1_3_mean", "win1_4_mean", "win1_5_mean", "win1_6_mean"]].apply(find_first_non_zero, axis=1)
    dp["win1_mean"] = all_competition[idx]["win1_mean"].astype(float)
    dp["win2_mean"] = all_competition[idx]["win2_mean"].astype(float)
    dp.dropna(inplace=True)
```

```python
data_col = train_data[0].columns
all_train_data = []
max_lenght = 0
for idx, dp in enumerate(train_data):
    all_train_data.append( torch.from_numpy(dp.drop('win2_mean', axis=1).copy().values).to(torch.float32) )
    
    max_lenght = max( all_train_data[-1].shape[0], max_lenght)
    
    p2_data_train = dp.drop('win1_mean', axis=1).dropna().copy()
    p2_data_train.loc[:, p2_data_train.columns != 'win2_mean'] *= -1
    all_competition[idx].loc[:, ["win2_2_mean","win2_3_mean", "win2_4_mean", "win2_5_mean", "win2_6_mean"]] = (all_freWIn[idx][["win2_2_mean","win2_3_mean", "win2_4_mean", "win2_5_mean", "win2_6_mean"]].values.astype(int)*insprice)
    p2_data_train.loc[:, "all_win"] = all_competition[idx][["win2_2_mean","win2_3_mean", "win2_4_mean", "win2_5_mean", "win2_6_mean"]].apply(find_first_non_zero, axis=1)
    all_train_data.append( torch.from_numpy(p2_data_train.dropna().values).to(torch.float32))
    
    max_lenght = max( all_train_data[-1].shape[0], max_lenght)
    
dataset_list = []
for ad in all_train_data:
    padded_tensor, mask, n = pad_and_create_mask(ad, max_lenght)
    dataset_list.append((padded_tensor,mask,n))
```

```python
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__(self, data_list):
        """
        Args:
            data_list (list of tuples): A list where each tuple contains
                                        (feature, mask), with
                                        - feature: Tensor of features
                                        - mask: Tensor of mask values
        """
        self.data_list = data_list

    def __len__(self):
        return len(self.data_list)

    def __getitem__(self, idx):
        feature, mask, n = self.data_list[idx]
        return feature, mask
    
    
@torch.no_grad()
def test_val( model, val_data ):
    acc = 0
    r_squared = 0
    for data, mask, n in val_data:
        data = data.unsqueeze(0)
        data = data.permute(1,0,2).to("cuda:0")
        outputs,_,_ = model(memory=data[:-1], tgt=data[1:],
                        tgt_mask=generate_square_subsequent_mask(len(data[1:])).to(data[1:].device))
        outputs = outputs.permute(1,0,2)[0,:n,-1].cpu()
        val_data = data[1:].permute(1,0,2)[0,:n,-1].cpu()
        
        ss_res = np.sum((outputs.numpy() - val_data.numpy()) ** 2)  # 残差平方和
        ss_tot = np.sum((outputs.numpy() - np.mean(val_data.numpy())) ** 2)  # 总平方和
        r_squared += 1 - (ss_res / ss_tot)
        
        res = outputs > 0.5
        val = val_data > 0.5
        
        acc += (res==val).sum()/val.shape[0]
        
    
    return acc, r_squared
        

train_dataset_list = dataset_list[:int(len(dataset_list)*0.8)]
val_dataset_list = dataset_list[int(len(dataset_list)*0.8):]
# 创建 CustomDataset 实例
dataset = CustomDataset(train_dataset_list)
# 使用 DataLoader 进行批量处理和数据打乱等
data_loader = DataLoader(dataset, batch_size=3, shuffle=True)
```

```python
model = TransformerDecoderLayer(12, 4, 100, 0.1, normalize_before=True).to("cuda:0")
model.train()

num_epochs = 100
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)


for epoch in range(num_epochs):  # num_epochs 是你希望训练的轮数
    for ad, mask in data_loader:  # 假设train_loader是你的数据加载器
        ad = ad.permute(1,0,2).to("cuda:0")

        # 前向传播
        outputs, _, _ = model(memory=ad[:-1], tgt=ad[1:],
                        tgt_mask=generate_square_subsequent_mask(len(ad[1:])).to(ad[1:].device))
        loss = criterion(outputs, ad[1:])
        
        acc, r2 = test_val(model, val_dataset_list)
        print(f"acc:{acc/len(val_dataset_list)}  ---------  r2:{r2/len(val_dataset_list)}")
        
        # 反向传播和优化
        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 0.1)
        optimizer.step()
```
